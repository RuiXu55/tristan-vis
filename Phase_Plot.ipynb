```python
def Phase_Plot(ind):
    pval = R.readprtl('ue',ind)
    xval = R.readprtl('xe',ind)/R.readparam('c_omp')
    #mycmap = new_cmaps.cmaps['viridis']
    zval = R.readflds('bx',ind)

    xmax = zval.shape[1]/R.readparam('c_omp')*R.readparam('istep')
    xmin = (xmax-1000)*0
   
    hist2d = np.histogram2d(pval,xval,bins = [1000,1500])
    cax = plt.imshow(hist2d[0],cmap=new_cmaps.cmaps['viridis'],norm=LogNorm(),
        origin='lower',aspect='auto',interpolation='hermite',
        extent=[0,xmax,min(pval),max(pval)])

    plt.xlabel(r'$x[c/\omega_{pe}]$')
    plt.ylabel(r'$\gamma\beta$')
    plt.xticks(np.linspace(xmin,xmax,5))
    plt.xlim([xmin,xmax])
    #plt.ylim([-0.4,0.2])
    plt.colorbar()
```

```python
'''
Make a single plot
'''
ind = 10
Phase_Plot(ind)
```

```python
def CreateMovie(plotter, fname, numberOfFrames, Interval=1, StartInd=1, fps=5):
    for i in range(numberOfFrames):
        print('Movie making index: i=',i)
        plotter(i*Interval+StartInd)
        plt.savefig(fname+str(i).zfill(3)+'.png',format='png')
        plt.clf()
    #os.system("rm movie.mp4")
    os.system("ffmpeg -framerate "+str(fps)+" -i "+fname+"%03d.png "+"../"+fname+".mp4")
    os.system("rm *.png")
```

```python
'''
Create a movie 
'''
fname = 'phase'
CreateMovie(Phase_Plot, fname, numberOfFrames=29, Interval=10, StartInd=1, fps=5)
```
